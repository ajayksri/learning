```python
# Modules

""" 
1. A module is simply a .py file. But not all modules are present in .py file. For example, some modules are built in
other languages.
2. Three ways to import modules
import mod1
import mod1, mod2, mod3
import mod1 as renameMod

3. import * means everything except private object which starts with "__".
4. Using __all__ in a module, the list of objects can be specified for importing.

5. How does python know where to look for modules and packages ?
    These are looked into sys.path first, then PYTHONPATH environment variable and finally from standard library
    present in python installable path.
    
6. Imports can quickly verified using "python -c" option.
7. When python file is executed -
    First it look for .pyo file (optimized byte code) [-O option or PYTHONOPTIMIZE environment variable]
    Second it look for .pyc (compiled byte code)
    -- Saving byte code can be avoided by -B option or PYTHONDONTWRITEBYTECODE environment variable
    
8. Whenever a module is imported Python creates a variable for the module called __name__ and stores the module’s name 
in this variable. A module’s name is simply the name of its .py ﬁle but without the extension. But whenever a .py ﬁle 
is run Python creates a variable for the program called __name__ and sets it to the string "__main__".
So a module test code could be written something like following -
if __name__ == "__main__": 
    import doctest 
    doctest.testmod()

"""
```
